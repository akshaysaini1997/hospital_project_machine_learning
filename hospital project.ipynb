```python
df=pd.DataFrame(df2,columns=['index', 'Patient Id', 'Age', 'Gender', 'Air Pollution', 'Alcohol use',
       'Dust Allergy', 'chronic Lung Disease', 'Smoking', 'Chest Pain',
       'Coughing of Blood', 'Shortness of Breath', 'Dry Cough',
       'Length Of Stay', 'Level','glucose', 'pulse'])
df
```

```python
df = df.drop(columns=['index', 'Patient Id'])
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = ['Level'])
y=df['Level']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
lb = LabelEncoder()
y_train_label = lb.fit_transform( y_train)
y_train_label
y_test_label = lb.fit_transform( y_test)
y_test_label
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn import svm
```

```python
linear = LinearRegression()
```

```python
linear.fit(x_train , y_train_label)
y_pred = linear.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test_label, y_pred)
```
